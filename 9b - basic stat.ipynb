```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Q9_b.csv',index_col=0)
data
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data['WT'], color='skyblue', edgecolor='black')
plt.title('Chick Weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```
